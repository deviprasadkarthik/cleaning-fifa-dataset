```python
for index,row in fifa.iterrows():
    if "On Loan" in row["Contract"] or "Free" in row["Contract"]:
        print(row['Contract'])
```

```python
def con_info(contract):
    if contract == "Free" or "On Loan" in contract:
        start_date=np.nan
        end_date=np.nan
        contract_len=0
    else:
        start_date,end_date= contract.split(" ~ ")
        start_year=int(start_date[:4])
        end_year=int(end_date[:4])
        contract_len=end_year-start_year
        
    return start_date,end_date,contract_len

#fun is applied to contarct col and creates new cols


new_cols=["contract start","contract end","contract length(years)"]
new_data=fifa["Contract"].apply(lambda x:pd.Series(con_info(x)))


for i in range(len(new_cols)):
    fifa.insert(loc=fifa.columns.get_loc('Contract')+1+i,column=new_cols[i] ,value=new_data[i])
```

```python
def contract_category(contract):
    if contract=="Free":
        return "Free"
    elif "On Loan" in contract:
        return "On Loan"
    else:
        "Contract"
        
        
fifa.insert(fifa.columns.get_loc("contract length(years)")+1, "contract status",fifa["Contract"].apply(contract_category))
        
    
        
```

```python
def con_hei(height):
    if "cm" in height:
        return int(height.strip("cm"))
    else:
        feet,inches=height.split("'")
        total_inch=int(feet)*12+int(inches.strip('"'))
        return round(total_inch*2.54)
fifa["Height"]=fifa["Height"].apply(con_hei)
fifa["Height"].unique()
        
        
```

```python
def convert_weight(weight):
    if "kg" in weight:
        return int(weight.strip("kg"))
    else:
        pounds=int(weight.strip("lbs"))
        return  (pounds/2.205)
    
fifa["Weight"]=fifa["Weight"].apply(convert_weight)
    
```
